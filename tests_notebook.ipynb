```python
import georef2
from pyexiv2 import Image
import numpy as np
import os
```

Testing distance between images by coverting GPS -> ECEF Cartesian Coordinates

```python
image_l = sorted(os.listdir("DJI_202508081433_021_PineIslandbog5H3m5x3photo"))
error_l = np.array([])
for i in range(13):
    path1 = os.path.join("DJI_202508081433_021_PineIslandbog5H3m5x3photo", image_l[i])
    path2 = os.path.join("DJI_202508081433_021_PineIslandbog5H3m5x3photo", image_l[i+1])
    img1 = Image(path1)
    img2= Image(path2)

    gps1 = float(img1.read_xmp()['Xmp.drone-dji.GpsLatitude']), float(img1.read_xmp()['Xmp.drone-dji.GpsLongitude'])
    gps2 = float(img2.read_xmp()['Xmp.drone-dji.GpsLatitude']), float(img2.read_xmp()['Xmp.drone-dji.GpsLongitude']) 

    cart1 = georef2.GPS_to_Cartesian(gps1)
    cart2 = georef2.GPS_to_Cartesian(gps2)

    distance = np.sqrt((cart2[0]-cart1[0])**2 + (cart2[1]-cart1[1])**2)
    str1 = image_l[i][-7:-4]
    str2 = image_l[i+1][-7:-4]
 
    error = abs(distance - 3)
    error_l = np.append(error_l, error)

print(error_l)
mean_error = np.mean(error_l)
print("max error: ", np.max(error_l))
print("mean error: ", mean_error)
```
